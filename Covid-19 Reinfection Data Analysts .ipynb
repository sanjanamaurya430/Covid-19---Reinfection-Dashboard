## KPI's Requirment

```python
# Total Covid Case
total_covid_case = df['Patient_ID'].count()

# Ventilator Rate
ventilator_rate = (df['Ventilator_Support'].str.upper() == 'YES').mean()

# Recovery Rate
recovery_rate = (df['Recovered'].str.upper() == 'YES').mean()

# ICU Admission Rate
icu_admission_rate = (df['ICU_Admission'].str.upper() == 'YES').mean()

# Display
print(f"Recovery Rate: {recovery_rate:.0%}")
print(f"Total Covid Case: {total_covid_case:,}")
print(f"Ventilator Rate: {ventilator_rate:.0%}")
print(f"ICU Admission Rate: {icu_admission_rate:.0%}")
```

## Reinfection by Age

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date_of_Infection' column to datetime format
df["Date_of_Infection"] = pd.to_datetime(df["Date_of_Infection"])

# Plot daily infection trend
plt.figure(figsize=(12,6))
df.groupby("Date_of_Infection")["Severity"].count().plot(kind="line", marker='o', color="blue")
plt.xlabel("Date of Infection")
plt.ylabel("COVID Cases")
plt.title("COVID-19 Daily Case Trends")
plt.xticks(rotation=45)
plt.show()
```

```python
#Visulization 2 : BMI by Severity
severity = df.groupby('Severity')['BMI'].mean()

plt.pie(severity,labels=severity.index,
      autopct='%.1f%%',
      startangle=90)

plt.title('BMI by Severity')
plt.axis('equal')
plt.show()
```

```python
# Visualization 3: Count plot for Gender and Reinfection
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Reinfection')
plt.title('Distribution of Reinfection')

plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Histogram for Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', bins=20, kde=True, color='orange')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 5 :Reinfection vs. Severity scatter plot
plt.figure(figsize=(10,5))
sns.scatterplot(x=df["Reinfection"], y=df["Severity"], hue=df["Vaccination_Status"], palette="coolwarm")
plt.xlabel("Reinfection Count")
plt.ylabel("Severity Level")
plt.title("COVID Reinfection vs. Severity Levels")
plt.show()
```

```python
# Visualization 6: Box Plot for BMI by Reinfection status
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Reinfection', y='BMI')
plt.title('BMI Distribution by Reinfection Status')
plt.xlabel('Reinfection')
plt.ylabel('BMI')
plt.show()
```
